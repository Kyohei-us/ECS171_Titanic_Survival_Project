# Filling null data

```python
# Finding missing data
miss_total = titanic.isnull().sum().sort_values(ascending=False)
percentage = miss_total/891 
percent_2 = (round(percentage, 3)).sort_values(ascending=False)
miss_total = pd.concat([miss_total, percent_2], axis=1, keys=['# of Null', 'Percentage'])
miss_total
```

```python
# drop Cabin column since it has huge amount of missing data, and it does not affect suvival (TODO: discuss this? Is there more about this attribute?)
titanic = titanic.drop(['Cabin'], axis=1)
# titanic.Cabin.unique()
```

```python
titanic
```

```python
titanic['Age'].plot.hist(grid=True, bins=20, rwidth=0.9,
                   color='#607c8e')
plt.title('Age disdribution before filling missing values')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
```

```python
titanic['Age'].mean()
```

```python
# Replace missing age by average(Age)
titanic['Age']=titanic['Age'].fillna(titanic['Age'].mean())
miss_total = titanic.isnull().sum().sort_values(ascending=False)
miss_total
```

```python
titanic['Age'].plot.hist(grid=True, bins=20, rwidth=0.9,
                   color='#607c8e')
plt.title('Age disdribution after filling missing values')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
```

```python
# Finally, the two missing value of embarked 
titanic['Embarked'].describe()
```

```python
# replace missing value with the most frequent one: S
titanic['Embarked'] = titanic['Embarked'].fillna('S')
```

```python
# Now no missing values!
miss_total = titanic.isnull().sum().sort_values(ascending=False)
miss_total
```

```python
titanic
```

# Encoding

```python
# Male=0, Female=1
titanic.loc[titanic['Sex']=='male','Sex']= 0
titanic.loc[titanic['Sex']=='female','Sex']= 1
titanic
```

```python
titanic['Ticket'].describe()
```

```python
# Drop Ticket column
titanic = titanic.drop(['Ticket'], axis=1)
```

```python
titanic.describe()
```

```python
titanic.info()
```

```python
# encode the Embarked column
titanic.loc[titanic['Embarked']=='S','Embarked']=0
titanic.loc[titanic['Embarked']=='C','Embarked']=1
titanic.loc[titanic['Embarked']=='Q','Embarked']=2
titanic
```

```python
titanic.info()
```

```python
# convert object types to int
titanic['Sex'] = titanic['Sex'].astype(int)
titanic['Embarked'] = titanic['Embarked'].astype(int)
```
